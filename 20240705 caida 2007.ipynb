```python
import pandas as pd
import pyarrow.parquet as pq
```

```python
# Specify the path to the Parquet file
parquet_file = 'Caida 2007\ddostrace.to-victim.20070804_134936.pcap output_dataframe.parquet'
# parquet_file = 'Caida 2007\ddostrace.to-victim.20070804_135436.pcap output_dataframe.parquet'

# Read Parquet file into a PyArrow Table
table = pq.read_table(parquet_file)

# Convert PyArrow Table to pandas DataFrame
df = table.to_pandas()
```
