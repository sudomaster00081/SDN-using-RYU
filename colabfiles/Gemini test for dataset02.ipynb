```python

def to_markdown(text):
    text = text.replace('•', '  *')
    return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))

def gemini_completion(prompt):
    """
    Generates text completion using the Gemini model from Google Generative AI.

    Args:
        prompt (str): The prompt text to generate completion.

    Returns:
        str or None: The generated text completion or None if an error occurs.
    """
    try:
        model_name = 'gemini-1.5-flash'  # Replace with the actual Gemini model name
        model = genai.GenerativeModel(model_name)

        response = model.generate_content(prompt)

        if response:
            return response.text
        else:
            raise ValueError("Empty response from Gemini model.")

    except ValueError as ve:
        print(f"ValueError occurred: {ve}")
    except Exception as e:
        print(f"Error occurred: {e}")

    return None

```

```python
import csv
def main():
  try :
    # CSV file path
    csv_file = 'data.csv'

    # Data for the new row
    new_row = [gemini_completion(prompt)]  # Example data, modify as needed
    print(new_row)

    # Open the CSV file in append mode
    with open(csv_file, mode='a', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(new_row)

    print(f'New row added to the CSV file "{csv_file}" successfully.')
    return 1
  except Exception as e:
    print(f'Error occurred: {e}')
    return 1
```

```python
for i in range(100):
  print(f"{i} th one in progress ")
  time.sleep(2)
  check = main()
  if check == 1:
    print(f"{i} th one done ")
    check = main()
  else:
    print(f"{i} th one failed ")
    check = main()
```
