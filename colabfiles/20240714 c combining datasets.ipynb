```python
# read data 1

df1 = pd.read_csv("data1.csv")
df1
```

```python
# read data 2

df2 = pd.read_csv("data2.csv")
df2
```

```python
# add columns as input and output for dfs
df1_with_heading = df1
df1_with_heading.columns = ["input", "output"]
df1_with_heading

df2_with_heading = df2
df2_with_heading.columns = ["input", "output"]
df2_with_heading
```

```python
#check if both the dataframe has same columns, if true append them
if list(df1_with_heading.columns) == list(df2_with_heading.columns):
  df_combined = pd.concat([df1_with_heading, df2_with_heading], ignore_index=True)
df_combined
```

```python
# remove duplicates in combined df
df_combined.drop_duplicates(subset=["input", "output"], keep="first", inplace=True)
df_combined
```
