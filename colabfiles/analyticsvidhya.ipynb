```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

import pandas as pd

# Minisom library and module is used for performing Self Organizing Maps

from minisom import MiniSom
# Loading Data
```

```python
# Load the dataset
url = "https://raw.githubusercontent.com/sudomaster00081/MscProj/main/Implimentation/combined%20Dataset.csv"
data = pd.read_csv(url)
```

```python
# X

data

# Shape of the data:

data.shape

# Info of the data:

data.info()
```

```python
# Defining X variables for the input of SOM
X = data.iloc[:, 0:4].values
y = data.iloc[:, -1].values
# X variables:
pd.DataFrame(X)
```

```python
pd.DataFrame(y)
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range = (0, 1))
X = sc.fit_transform(X)
pd.DataFrame(X)
```

```python
# Set the hyper parameters
som_grid_rows = 10
som_grid_columns = 10
iterations = 20000
sigma = 1
learning_rate = 0.5
```

```python
X
```

```python
# define SOM:

som = MiniSom(x = som_grid_rows, y = som_grid_columns, input_len=4, sigma=sigma, learning_rate=learning_rate)

# Initializing the weights

som.random_weights_init(X)

# Training

som.train_random(X, iterations)
```

```python
# Weights are:
wts = som.weights
# Shape of the weight are:
wts.shape
# Returns the distance map from the weights:
som.distance_map()
```
